```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data for each algorithm
ppo_results = pd.read_csv('logs/ppo/PongNoFrameskip-v4_7/0.monitor.csv', skiprows=2, names=['r', 'l', 't'])
dqn_results = pd.read_csv('logs/dqn/PongNoFrameskip-v4_3/0.monitor.csv', skiprows=2, names=['r', 'l', 't'])
a2c_results = pd.read_csv('logs/a2c/PongNoFrameskip-v4_2/0.monitor.csv', skiprows=2, names=['r', 'l', 't'])

# Compute rolling mean for rewards and episode lengths
ppo_results['reward_mean'] = ppo_results['r'].rolling(window=50).mean()
ppo_results['episode_length'] = ppo_results['l'].rolling(window=50).mean()
dqn_results['reward_mean'] = dqn_results['r'].rolling(window=50).mean()
dqn_results['episode_length'] = dqn_results['l'].rolling(window=50).mean()
a2c_results['reward_mean'] = a2c_results['r'].rolling(window=50).mean()
a2c_results['episode_length'] = a2c_results['l'].rolling(window=50).mean()

# Plot the smoothed rewards
plt.plot(ppo_results['t'], ppo_results['reward_mean'], label='PPO')
plt.plot(dqn_results['t'], dqn_results['reward_mean'], label='DQN')
plt.plot(a2c_results['t'], a2c_results['reward_mean'], label='A2C')

# Add labels, title, and legend
plt.xlabel('Time (seconds)')
plt.ylabel('Average Reward')
plt.title('Comparison of Average Reward over Time during Training')
plt.legend()

# Display the plot
plt.show()

# Plot the smoothed episode lengths
plt.plot(ppo_results['t'], ppo_results['episode_length'], label='PPO')
plt.plot(dqn_results['t'], dqn_results['episode_length'], label='DQN')
plt.plot(a2c_results['t'], a2c_results['episode_length'], label='A2C')

# Add labels, title, and legend
plt.xlabel('Time (seconds)')
plt.ylabel('Episode Length')
plt.title('Comparison of Average Episode Length over Time during Training')
plt.legend()

# Display the plot
plt.show()
```

```python
#Calculating the mean reward during training for each algorithm
ppo_mean_reward = ppo_results['r'].mean()
print(f"The mean overall reward for ppo is: {ppo_mean_reward}")

dqn_mean_reward = dqn_results['r'].mean()
print(f"The mean overall reward for dqn is: {dqn_mean_reward}")

a2c_mean_reward = a2c_results['r'].mean()
print(f"The mean overall reward for a2c is: {a2c_mean_reward}")
```

```python
#Calculating the mean episode length during training for each algorithm
ppo_mean_length = ppo_results['l'].mean()
print(f"The mean overall episode length for ppo is: {ppo_mean_length}")

dqn_mean_length = dqn_results['l'].mean()
print(f"The mean overall episode length for dqn is: {dqn_mean_length}")

a2c_mean_length = a2c_results['l'].mean()
print(f"The mean overall episode length for a2c is: {a2c_mean_length}")
```
